```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga el dataset 'tips' de ejemplo que viene con Seaborn y muestra las. NO necesario externo.
tips = sns.load_dataset('tips')
tips.head()
```

**Paletas de colores**

https://seaborn.pydata.org/tutorial/color_palettes.html

con esto podemos ver los diferentes colores q podemos aplicar.

```python
sns.relplot(x='total_bill', y='tip', data=tips);
# Crea un diagrama de dispersión (scatterplot) que muestra la relación entre 'total_bill' (cuenta total) y 'tip' (propina) del dataset 'tips'.
```

```python
sns.set_theme(style='darkgrid')#cambiar la forma
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='time', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='size', data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', palette="pastel", data=tips);
```

Tamaños

```python
sns.relplot(x='total_bill', y='tip', size='size',  data=tips);
#CAmbien el tamaño segun marcador, muxos comensales grande, pequeños pequeña.
```

```python
sns.relplot(x='total_bill', y='tip', size='size', sizes=(20, 150), data=tips[tips['size'] < 4])
# Crea un diagrama de dispersión que relaciona 'total_bill' y 'tip'.
# El tamaño de los puntos varía según la columna 'size' (número de personas en la mesa).
# 'sizes=(20, 150)' establece el rango de los tamaños de los puntos.
# 'data=tips[tips['size'] < 4]' filtra el dataset para mostrar solo las filas donde 'size' es menor a 4.
```

# Lineplot

```python
df=pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
sns.relplot(x='time', y='value', kind='line', data=df)
# Crea un DataFrame con una columna 'time' (de 0 a 499) y una columna 'value' (valores aleatorios acumulados).
# Genera un gráfico de líneas que muestra la evolución de 'value' en función de 'time'. 'kind='line'' especifica que se quiere un gráfico de líneas.
# Es aleatorio
```

```python
df=pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=['x', 'y'])
sns.relplot(x='x', y='y',sort= False, kind="line", data=df)
# Crea un DataFrame con dos columnas ('x' e 'y') que representan la acumulación de números aleatorios.
# Genera un gráfico de líneas que muestra la relación entre 'x' e 'y'.
# 'sort=False' desactiva la ordenación de los datos antes de dibujar la línea. Esto puede ser útil si los datos ya están ordenados o si quieres preservar el orden original.
# 'kind="line"' especifica que se quiere un gráfico de líneas.
```
