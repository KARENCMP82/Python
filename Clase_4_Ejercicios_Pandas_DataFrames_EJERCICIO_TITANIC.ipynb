2. Identifique qué columnas tienen valores nulos


```python
#Validación de valores nulos
# Filtrar columnas con valores nulos
columnas_con_nulos = titanic.isnull().sum()[titanic.isnull().sum() > 0]
# etiqueta               buscar nulos/ sum   Esto seleciona solo  las filas de nulos
print(columnas_con_nulos) #muestra por pantalla
```

3. Sustituya los valores nulos en la columna "Cabin" por el valor más frecuente (moda)

```python
#Imputar valores nulos: #Cabin: por valores mas frecuentes, son ..B96 B98
titaCopi['Cabin'] = titanic['Cabin'].fillna(titanic['Cabin'].mode()[0])
#     obtiene valores frecuentes  /  rellena valores null de la columna
```

4. Elimine las filas donde la columna 'Embarked' está nula.


```python
titaCopi = titaCopi.dropna(subset=['Embarked'])
#dropna  eliminar las filas con valores faltantes. subset dice q columna
```

```python
columnas_con_nulos = titaCopi.isnull().sum()[titaCopi.isnull().sum() > 0]
print(columnas_con_nulos)
```

7. Mostrar por pantalla las filas pares del DataFrame.

```python
filas_pares = titaCopi.iloc[::2]
print(filas_pares)
```

[::2]: Este es un operador de rebanado que selecciona elementos de una secuencia (en este caso, las filas del DataFrame).

El primer : indica que se seleccionarán todas las filas.

El segundo : indica que se seleccionarán todas las columnas.

El 2 indica que se tomará cada segunda fila, comenzando desde la primera (índice 0), lo cual resulta en las filas pares.

8. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.

```python
# Filtrar los pasajeros de primera clase y ordenar por nombre
nombres_primera_clase = titaCopi[titaCopi.Pclass == 1]['Name'].sort_values()

# Mostrar los nombres
print(nombres_primera_clase)
```

```python
titaCopi2 = titaCopi.copy()
```

**Cambiar nombre de columnas**

```python
titaCopi2.head()
```

```python
#Cambiar nombre de columnas
titaCopi2.rename(columns={
    "PassengerId": "Id_pasaj",
    "Survived": "Sobrevivi",
    "Pclass": "Clase",
    "Name": "Nombre",
    "Sex": "Sexo",
    "Age": "Edad",
    "SibSp": "Hermanos/esposos_a_bordo",
    "Parch": "Padres/hijos_a_bordo",
    "Ticket": "Billete",
    "Fare": "Tarifa",
    "Cabin": "Cabina",
    "Embarked": "Embarcado"
}, inplace=True)


```

9. Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.


```python
# Calcular el número de personas que sobrevivieron y murieron
conteo_sobrevivientes = titaCopi2["Sobrevivi"].value_counts()

# Calcular el porcentaje / nº total de filas len(titaCopi2)
porcentaje_sobrevivientes = conteo_sobrevivientes / len(titaCopi2) * 100

# Mostrar los porcentajes
print(porcentaje_sobrevivientes)
```

10. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.

```python
# Filtrar los pasajeros por clase
primera_clase = titaCopi2[titaCopi2.Clase == 1]
segunda_clase = titaCopi2[titaCopi2.Clase == 2]
tercera_clase = titaCopi2[titaCopi2.Clase == 3]
```

```python
# Calcular el porcentaje de sobrevivientes por clase
porcSupPrimer = primera_clase["Sobrevivi"].mean() * 100
porcSupSegunda = segunda_clase["Sobrevivi"].mean() * 100
porcSupTercera = tercera_clase["Sobrevivi"].mean() * 100
```

```python
# Mostrar los porcentajes
print(f"Porcentaje de sobrevivientes en primera clase: {porcSupPrimer:.2f}%")
print(f"Porcentaje de sobrevivientes en segunda clase: {porcSupSegunda:.2f}%")
print(f"Porcentaje de sobrevivientes en tercera clase: {porcSupTercera:.2f}%")
```

11. Eliminar del DataFrame los pasajeros con edad desconocida.

```python
titaCopi3 = titaCopi2.copy()
```

```python
# Eliminar nulos dropna, subset especifica la columna
titaCopi3 = titaCopi3.dropna(subset=['Edad'])
```

12. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

```python
titaCopi3.head()
```

```python

# Filtrar las filas donde el sexo es femenino y luego calcular la media de la columna 'Edad'
media = titaCopi3[titaCopi3['Sexo'] == 'female']['Edad'].mean()
print(media)
```

13. Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.

```python
titaCopi3['MayorDeEdad'] = titaCopi3['Edad'] > 18
```

```python
titaCopi3.head()
```

14.Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.

```python
# Separar los mayores de edad y los menores de edad
PorcentMayor = titaCopi3[titaCopi3['MayorDeEdad'] == True]
PorcentMenor = titaCopi3[titaCopi3['MayorDeEdad'] == False]

# Calcular y mostrar el porcentaje de sobrevivientes para mayores y menores en cada clase
for clase in [1, 2, 3]:
    # Filtrar por clase
    mayores_clase = PorcentMayor[PorcentMayor['Clase'] == clase]
    menores_clase = PorcentMenor[PorcentMenor['Clase'] == clase]

    # Calcular los porcentajes
    ResutMayo = mayores_clase['Sobrevivi'].mean() * 100 if not mayores_clase.empty else 0
    ResutMenor = menores_clase['Sobrevivi'].mean() * 100 if not menores_clase.empty else 0

    # Mostrar resultados
    print(f"Clase {clase}:")
    print(f"  Porcentaje de mayores que sobrevivieron: {ResutMayo:.2f}%")
    print(f"  Porcentaje de menores que sobrevivieron: {ResutMenor:.2f}%")


```
