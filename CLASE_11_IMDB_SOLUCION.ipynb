### 3.3.1 Tratamiento de outliers

```python
imdb3[imdb3["duration"]>300]
# vemos duracion a 300min
```

Hay varias formas de reemplazarlos

### 3.3.1.1 Primera opción: funciones *iloc* y *loc* de pandas

iloc[i, j] : Devuelve el elemento que se encuentra en la fila i y la columna j del DataFrame

```python
imdb3.iloc[6,:]
  #selecciona una fila específica (la número 6, empezando desde 0)
```

loc[] es un método para seleccionar y filtrar datos en un DataFrame de Pandas. El método loc[] toma uno o dos argumentos, que pueden ser una etiqueta de fila o una matriz booleana.

```python
# Sustituye los valores en la columna "duration" que son menores a 50 o mayores a 300
# con la media (mean) de la columna "duration". Esto ayuda a manejar valores atípicos.
imdb3.loc[(imdb3["duration"] < 50) | (imdb3["duration"] > 300), 'duration'] = imdb3["duration"].mean()
```

```python
imdb3.hist("duration")
```

### 3.3.1.2 Segunda opción: función *np.where* de numpy

```python
imdb3 = imdb2.copy()
#recuperamos la imd2 para 3
```

```python
# Reemplaza los valores de la columna "duration" que sean menores o iguales a 50
# con la media de la columna "duration". El resto de los valores permanecen iguales.
imdb3['duration'] = np.where(imdb3['duration'] <= 50, imdb3["duration"].mean(), imdb3['duration'])
```

```python
# Reemplaza los valores de la columna "duration" mayores a 300
# con la media de la columna "duration". El resto de los valores permanecen iguales.
imdb3['duration'] = np.where(imdb3['duration'] > 300, imdb3["duration"].mean(), imdb3['duration'])
```

```python
imdb3.hist("duration")
```

### 3.3.1.3 Tercera opción: función *drop* de pandas

```python
imdb3 = imdb2.copy()
```

```python
# Obtiene los índices de las filas donde la columna "duration" tiene valores
# menores a 50 o mayores a 300.
imdb3[(imdb3["duration"] < 50) | (imdb3["duration"] > 300)].index
```

```python
# Calcula el número total de filas en el DataFrame "imdb3".
len(imdb3)
```

```python
# Elimina del DataFrame "imdb3" todas las filas donde la columna "duration"
# sea menor a 50 o mayor a 300. La operación se realiza directamente en el DataFrame original.
imdb3.drop(imdb3[(imdb3["duration"] < 50) | (imdb3["duration"] > 300)].index, inplace=True)
```

```python
# Calcula el número total de filas en el DataFrame "imdb3".
len(imdb3)
```

### 3.3.2 Tratamiento de la variable género

```python
# Divide los valores de la columna "genres" del DataFrame "imdb3" en varias columnas,
# separando por el carácter "|". Cada género individual queda en una columna diferente.
# El resultado se guarda en un nuevo DataFrame llamado "genero_list".
genero_list = imdb3["genres"].str.split("|", expand=True)
```

```python
genero_list[0]  # Primera columna (género principal de cada película)
```

```python
genero_list.fillna('sin genero', inplace=True)
# Rellena los valores faltantes (NaN) en la lista 'genero_list' con la cadena 'sin genero'.
# Modifica la lista original directamente.
```

```python
for i in genero_list:
    # Recorre cada género en la lista 'genero_list'
    imdb3["genero_"+str(i)]=genero_list[i]
    # Crea una nueva columna en 'imdb3' con el nombre 'genero_' seguido del género actual
    # Asigna el valor del género actual a todos los registros de la nueva columna
```
